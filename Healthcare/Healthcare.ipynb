```python
import pandas as pd
import matplotlib.pylab as plt 
import plotly.express as px
```

```python
file_path = "healthcare_dataset.csv"
df = pd.read_csv(file_path)

# Remove NA
df_clean = df[df['Name'].notna()]

gender_count = df_clean['Gender'].value_counts()
bloodType_count = df_clean['Blood Type'].value_counts
condition_count = df_clean['Medical Condition'].value_counts

# Condition vs Gender
condition_gender = df_clean.groupby(['Gender', 'Medical Condition']).size().unstack(fill_value = 0)

# Plotting
plt.figure(figsize=(10, 6))
condition_gender.plot(kind='bar', stacked= True)

# Adding labels and title
plt.title('Gender vs Medical Condition')
plt.xlabel('Gender')
plt.ylabel('Count of Medical Conditions')
plt.xticks(rotation=0)  # Rotate x labels for better readability
plt.legend(title='Medical Condition')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
total_patients = df_clean['Name'].nunique()
total_bill = df_clean['Billing Amount'].sum()
total_hospital = df_clean['Hospital'].nunique()
total_doctor = df_clean['Doctor'].nunique()
total_insurance = df_clean["Insurance Provider"].nunique()
```

```python
# Distribution of Insurance  Provider
# Count occurrences of each Insurance Provider
insurance_counts = df_clean['Insurance Provider'].value_counts().reset_index()
insurance_counts.columns = ['Insurance Provider', 'Count']
fig1 = px.pie(insurance_counts, 
              names= 'Insurance Provider',
              values = 'Count',
              title = 'Patient Distribution by Insurance Provider',
              color = 'Count',
              color_discrete_sequence =px.colors.qualitative.Set3)
fig1.update_layout(
    width=700,  # Set the width of the pie chart
    height=500  # Set the height of the pie chart
)

fig1
```
